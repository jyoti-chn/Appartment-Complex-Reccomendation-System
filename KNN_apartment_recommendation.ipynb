```python
# using KNN for optimal solutions
import pandas as pd
import numpy as np
from sklearn.neighbors import NearestNeighbors

# fetch the normalized and cleaned data
data_encoded_normalised = pd.read_csv('/content/cleaned_housing_data_scaled_normalized.csv')
data_encoded = pd.read_csv('/content/cleaned_housing_data_scaled.csv')

# remove unnamed column
data_encoded = data_encoded.loc[:, ~data_encoded.columns.str.contains('^Unnamed')]
data_encoded_normalised = data_encoded_normalised.loc[:, ~data_encoded_normalised.columns.str.contains('^Unnamed')]

data_encoded_normalised.head()
```

```python
# load the KNN model
model=NearestNeighbors(metric='cosine')
model.fit(data_encoded_normalised)
```

```python
# user requirements, here we are selecting a random sample from the data as user reuirments:
selected=7


data_encoded_normalised.iloc[selected].values[0:8]
```

```python
#The model will give index of the 6 similar apartments as per user.
dist,index=model.kneighbors(data_encoded_normalised.iloc[selected].values.reshape(1,-1),n_neighbors=6)
index
```

```python
reccomendation=[]
for x in index:
    reccomendation.append(x)
reccomendation
```

```python
# printing the user requirments
r=data_encoded.iloc[[selected]][data_encoded.iloc[[selected]]>0].dropna(axis=1)
r
```

```python
# print the 6 similar recommendation for the user
col=r.columns
for x in index:
    suggestions=data_encoded.iloc[list(x)][col]
suggestions
```
